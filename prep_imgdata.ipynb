```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import itertools
import random
import pickle
from tempfile import TemporaryFile
from tqdm import tqdm
```

```python
    def __init__(self,datadir,img_sz, color = 'g'):
        self.datadir = datadir
        self.img_sz = img_sz
        self.color = color
        if color == 'g':
            TRAIN_COLOR = cv2.IMREAD_GRAYSCALE
        else:
            TRAIN_COLOR = cv2.IMREAD_COLOR
        self.fit_color = TRAIN_COLOR

    def show_resize_img(self, imgu_SIZE):

        if  imgu_SIZE != self.img_sz:
            IMG_SIZE = imgu_SIZE
        else:
            IMG_SIZE = self.img_sz
        TRAINDIR = "/home/howard/data/{}/train".format(self.datadir)
        categors = []
        for category in os.listdir(TRAINDIR):
                categors.append(category)
        CATEGORIES = categors
        for category in CATEGORIES:
            path = os.path.join(TRAINDIR, category)
        for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path,img), self.fit_color)
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))

        if self.color == 'c':
            RGB_img = cv2.cvtColor(new_array, cv2.COLOR_BGR2RGB)
            plt.imshow(RGB_img)
        else:
            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
            plt.imshow(new_array, cmap="gray")
            
        plt.show()
    def create_test_data(self):
        IMG_SIZE = self.img_sz


        TESTDIR = "/home/howard/data/{}/valid".format(self.datadir)
        training_data = []
        for img in os.listdir(TESTDIR):
            class_num = img[:-9]
            #print(class_num)
            try:
                img_array = cv2.imread(os.path.join(TESTDIR,img), self.fit_color)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                training_data.append([new_array, class_num])
            except Exception as e:
                        print('fail')
                        pass
        test_X = []
        test_Y = []
        for features,label in training_data:
            test_X.append(features)
            test_Y.append(label)
        test_X =  np.array(test_X).reshape(-1, IMG_SIZE,IMG_SIZE, 1)
        return test_X, test_Y
    def create_train_data(self):
        count = 1
        for i in tqdm(range(1)):
            print(count)
            count += 1
            TRAINDIR = "/home/howard/data/{}/train".format(self.datadir)
            categors = []
            testcategors = []
            X = []
            Y = []
            IMG_SIZE = self.img_sz
            for category in os.listdir(TRAINDIR):
                categors.append(category)
            CATEGORIES = categors
            #THIS IS THE CONVERT PATH IMAGES TO DATA
            for category in CATEGORIES:
                path = os.path.join(TRAINDIR, category)
                for img in os.listdir(path):
                    img_array = cv2.imread(os.path.join(path,img), self.fit_color)
            training_data = []
            class_nums = []
            numtolet = {}

            def create_training_data():
                for category in CATEGORIES:
                    path = os.path.join(TRAINDIR, category)
                    class_num = CATEGORIES.index(category)
                    class_nums.append(class_num)
                    for img in os.listdir(path):
                        try:
                            img_array = cv2.imread(os.path.join(path,img), self.fit_color)
                            new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                            training_data.append([new_array, class_num])
                        except Exception as e:
                            pass
            create_training_data()
            random.shuffle(training_data)
            numtolet = dict(zip(class_nums, CATEGORIES))
            for features,label in training_data:
                X.append(features)
                Y.append(label)
            X =  np.array(X).reshape(-1, IMG_SIZE,IMG_SIZE, 1)
            self.categories_len = len(numtolet)
            np.save('X-{}'.format(self.datadir), X)
            np.save('Y-{}'.format(self.datadir), Y)
            np.save('numtolet-{}'.format(self.datadir), numtolet)
            print("done")
            return X, Y, numtolet
#IMGPREP = img_prep()
#create_train_data(datadir='asl-alphabet',img_sz=100)
#IMGPREP = img_prep()
#create_train_data(datadir='asl-alphabet',img_sz=100)
```
